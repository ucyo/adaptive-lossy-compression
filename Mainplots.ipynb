```python
from run import get_corr
from matplotlib import pyplot as plt
from matplotlib import gridspec
%matplotlib inline
```

```python
%%capture
nao = get_corr('nao','mm',6,3,10)
qbo = get_corr('qbo30','mm',6,3,10)
```

```python
f = plt.figure()
gs = gridspec.GridSpec(2, 2, width_ratios=[2,1])

ax1 = plt.subplot(gs[0])
ax2 = plt.subplot(gs[1])
ax3 = plt.subplot(gs[2], sharex=ax1)
ax4 = plt.subplot(gs[3], sharex=ax2)


everyone = [x for x in nao.columns if x[:3]!='raw' and x!='truth']
focus = ['roll', 'special', 'zfp']

naoALL = nao.loc[2:,everyone].plot(figsize=(20,10), xlim=(2,70), ax=ax2, legend=False)
naoFOC = nao.loc[2:,focus].plot(figsize=(20,10), color=[x.get_color() for x in naoALL.lines[-3:]], ax=ax1, legend=False)
qboALL = qbo.loc[2:,everyone].plot(figsize=(20,10), xlim=(2,70), ax=ax4, legend=False)
qboFOC = qbo.loc[2:,focus].plot(figsize=(20,10), color=[x.get_color() for x in naoALL.lines[-3:]], ax=ax3, legend=False)

_ = ax3.set_xlabel('Time (mm)', fontsize=16)
_ = ax4.set_xlabel('Time (mm)', fontsize=16)
_ = ax3.set_ylabel('Pearson Corr. (QBO30)', fontsize=16)
_ = ax1.set_ylabel('Pearson Corr. (NAO)', fontsize=16)
_ = ax1.axvline(70, color='grey', linestyle='--')
_ = ax3.axvline(70, color='grey', linestyle='--')
_ = ax3.axvspan(0, 70, color='grey', alpha=0.05)
_ = ax1.axvspan(0, 70, color='grey', alpha=0.05)
_ = ax1.tick_params(axis='both', which='both', labelsize=14)
_ = ax2.tick_params(axis='both', which='both', labelsize=14)
_ = ax3.tick_params(axis='both', which='both', labelsize=14)
_ = ax4.tick_params(axis='both', which='both', labelsize=14)

handles, labels = ax4.get_legend_handles_labels()

# Label modification
corrections = dict(zfp='direct')
tmp = [x if x not in corrections.keys() else corrections[x] for x in labels]
labels = [x.upper() for x in tmp]

lgd = f.legend(handles=handles, labels=labels,
             loc='center', fontsize=17, ncol=7,
            bbox_to_anchor=(.5, .03),borderaxespad=0)
f.tight_layout()
plt.subplots_adjust(bottom=0.12)
plt.savefig('/home/ucyo/Downloads/pof/new.png', pad_inches=0.01)
```
